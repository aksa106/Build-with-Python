```python
# Sum of N Even Natural Numbers
# Problem Description:

# You are given an integer n. Your task is to calculate and return the sum of the first n even natural numbers. The even natural numbers are: 2, 4, 6, 8, ...



# Input:

# A single integer n where 1 <= n <= 10^4.



# Output:

# Return the sum of the first n even natural numbers.



# Example:

# Input: n = 3
# Output: 12  # (2 + 4 + 6)
 
# Input: n = 5
# Output: 30  # (2 + 4 + 6 + 8 + 10)
def sum_of_even_natural_numbers(n):
    total = 0
    even = 2
    count = 0

    while count < n :
        total += even 
        even += 2
        count += 1


    return total 

print(sum_of_even_natural_numbers(3))

print(sum_of_even_natural_numbers(5))

    
```

```python
## ODD VERSION 
def sum_of_odd_natural_numbers(n):
    i = 0
    total_sum = 0
    odd = 1
    
    while i < n :
        total_sum += odd
        odd += 2        ## moves to the next odd number
        i += 1

    return total_sum

print(sum_of_odd_natural_numbers(3))
```

```python
# Check for Prime Number
# Problem Description:

# You are given an integer n. Your task is to check whether the number is prime or not. A prime number is a number greater than 1 that has no divisors other than 1 and itself. Return True if the number is prime, and False otherwise.



# Input:

# A single integer n where 1 <= n <= 10^6.



# Output:

# Return True if n is a prime number, otherwise return False.



# Example:

# Input: n = 5
# Output: True
 
# Input: n = 4
# Output: False

def prime(n):
    if n <= 1:
        return False
    
    i = 2

    while i<n:
        if n % i == 0:
            return False
        
        i += 1

    return True

print(prime(5))
print(prime(4))
```
